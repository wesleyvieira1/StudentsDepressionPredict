````python
df = df.loc[df['Age'] <= 30]
df = df.loc[df['Academic Pressure'] > 0]
df = df.loc[df['Study Satisfaction'] > 0]
````

### 5. Encoder

**Estratégias aplicadas:**
- **Ordinais:** Sleep Duration, Dietary Habits, Age (ordem natural)
- **Binárias:** Gender, histórico familiar, pensamentos suicidas
- **Hierárquicas:** Degree (nível educacional crescente)

#### 5.1 Coluna 'Sleep Duration'

````python
def sleep_quality(sleep):
    if sleep == 'Less than 5 hours':
        return 1
    elif sleep == '5-6 hours':
        return 2
    elif sleep == '7-8 hours':
        return 3
    elif sleep == 'More than 8 hours':
        return 4
    else:
        return np.nan

df['Sleep Duration'] = df['Sleep Duration'].apply(sleep_quality)
````

#### 5.2 Coluna 'Dietary Habits'

````python
def dietary_habits(habit):
    if habit == 'Unhealthy':
        return 1
    if habit == 'Moderate':
        return 2
    if habit == 'Healthy':
        return 3
    else:
        return np.nan

df['Dietary Habits'] = df['Dietary Habits'].apply(dietary_habits)
df['Dietary Habits'].value_counts()
````

#### 5.4 Binarização 

````python
df['Gender'] = df['Gender'].map({'Female':1,'Male': 0})
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map({'Yes':1,'No': 0})
df['Family History of Mental Illness']= df['Family History of Mental Illness'].map({'Yes':1,'No':0})
````

#### 5.5 Coluna 'Degree'

1. **Class 12** (Ensino Médio ou equivalente)
2. **B.Ed**, **B.Com**, **B.Arch**, **BCA**, **B.Tech**, **BHM**, **BSc**, **B.Pharm**, **BBA**, **BE**, **BA** (Graduação - Bacharelado)
3. **MSc**, **MCA**, **M.Tech**, **M.Ed**, **M.Com**, **MBA**, **MA**, **M.Pharm**, **MHM**, **ME** (Pós-Graduação - Mestrado)
4. **MBBS**, **MD** (Medicina - Graduação e Pós-Graduação)
5. **LLB**, **LLM** (Direito - Graduação e Pós-Graduação)
6. **PhD** (Doutorado - Nível mais alto de educação acadêmica)

````python
def superity_level(degree):
    if degree == 'Class 12':
        return 1
    elif degree in ['B.Ed', 'B.Com', 'B.Arch', 'BCA', 'B.Tech', 'BHM', 'BSc', 'B.Pharm', 'BBA', 'BE', 'BA']:
        return 2
    elif degree in ['MSc', 'MCA', 'M.Tech', 'M.Ed', 'M.Com', 'MBA', 'MA', 'M.Pharm', 'MHM', 'ME']:
        return 3
    elif degree in ['MBBS', 'MD']:
        return 4
    elif degree in ['LLB', 'LLM']:
        return 5
    elif degree == 'PhD':
        return 6
    else:
        return np.nan

df['Degree'] = df['Degree'].apply(superity_level)
````

### 6. Feature Engineering

#### 6.1 Índice de Estresse Geral

A coluna `Weighted Stress` foi criada para representar uma métrica ponderada que combina os níveis de pressão acadêmica (`Academic Pressure`) e estresse financeiro (`Financial Stress`). Essa abordagem permite identificar indivíduos que enfrentam altos níveis de estresse em ambas as áreas, fornecendo uma base para análises mais detalhadas e intervenções direcionadas.
```

````python
# Definir os pesos para cada variável
weight_academic = 0.6  # Peso para Academic Pressure
weight_financial = 0.4  # Peso para Financial Stress

# Criar a nova métrica ponderada
df['Weighted Stress'] = (df['Academic Pressure'] * weight_academic + df['Financial Stress'] * weight_financial) / (weight_academic + weight_financial)
````

#### 6.2 Vulnerabilidade Mental

A coluna `Mental_Vulnerability_Score` foi criada para representar uma métrica composta que avalia a vulnerabilidade mental de um indivíduo com base em múltiplos fatores. Esses fatores incluem pensamentos suicidas, histórico familiar de doenças mentais, pressão acadêmica, estresse financeiro e qualidade do sono. Cada fator foi ponderado de acordo com sua relevância, permitindo uma análise mais detalhada do impacto combinado desses elementos na saúde mental. 

````python
df['Mental_Vulnerability_Score'] = (
    df['Have you ever had suicidal thoughts ?'] * 0.4 +
    df['Family History of Mental Illness'] * 0.2 +
    (df['Academic Pressure'] / 5) * 0.2 +
    (df['Financial Stress'] / 5) * 0.1 +
    ((5 - df['Sleep Duration']) / 4) * 0.1 
)
````

#### 6.3 Crise mental severa

A coluna `Extreme_Mental_Crisis` foi criada para identificar indivíduos que apresentam uma combinação de fatores críticos relacionados à saúde mental. Essa métrica binária (0 ou 1) é ativada quando o indivíduo relata pensamentos suicidas e, simultaneamente, apresenta pelo menos um dos seguintes fatores de risco em níveis elevados:

- Pressão acadêmica (`Academic Pressure`) maior ou igual a 4.
- Estresse financeiro (`Financial Stress`) maior ou igual a 4.
- Duração do sono (`Sleep Duration`) menor ou igual a 2.


````python
df['Extreme_Mental_Crisis'] = (
    (df['Have you ever had suicidal thoughts ?'] == 1) & 
    (
        (df['Academic Pressure'] >= 4) | 
        (df['Financial Stress'] >= 4) | 
        (df['Sleep Duration'] <= 2)
    )
).astype(int)
````

### 6.4 Depressão ultimato

A coluna `Ultimate_Depression_Predictor` é uma métrica composta que avalia a probabilidade de depressão com base em fatores de risco, como pensamentos suicidas, histórico familiar, pressão acadêmica e financeira, hábitos de sono e alimentares, satisfação com os estudos e carga horária, com pesos específicos para cada fator.

````python
df['Ultimate_Depression_Predictor'] = (
    df['Have you ever had suicidal thoughts ?'] * 0.35 +
    df['Family History of Mental Illness'] * 0.15 +
    ((df['Academic Pressure'] >= 4) & (df['Financial Stress'] >= 4)).astype(int) * 0.20 +
    ((df['Sleep Duration'] <= 2) & (df['Dietary Habits'] <= 2)).astype(int) * 0.15 +
    ((df['Study Satisfaction'] <= 2) & (df['Academic Pressure'] >= 4)).astype(int) * 0.10 +
    (df['Work/Study Hours'] >= 8).astype(int) * 0.05
).round(2)
````

### 6.5 Mental Health Crisis Index



A coluna `Mental_Health_Crisis_Index` avalia o risco de crise de saúde mental com base em fatores como pensamentos suicidas, pressão acadêmica, estresse financeiro, sono, hábitos alimentares, histórico familiar e carga horária, ponderados por relevância. 


````python
df['Mental_Health_Crisis_Index'] = (
    (df['Have you ever had suicidal thoughts ?'] == 1).astype(int) * 3 +
    ((df['Academic Pressure'] >= 4) & (df['Study Satisfaction'] <= 2)).astype(int) * 2 +
    (df['Financial Stress'] >= 4).astype(int) * 2 +
    ((df['Sleep Duration'] <= 2) | (df['Dietary Habits'] <= 2)).astype(int) * 1 +
    (df['Family History of Mental Illness'] == 1).astype(int) * 1 +
    (df['Work/Study Hours'] >= 10).astype(int) * 1
)
````

### 6.8 Stress Vulnerability Multiplier  


A coluna `Stress_Vulnerability_Multiplier` é uma métrica composta que avalia a vulnerabilidade ao estresse com base em múltiplos fatores, como pressão acadêmica, estresse financeiro, pensamentos suicidas, histórico familiar de doenças mentais, duração do sono e satisfação com os estudos.


````python
df['Stress_Vulnerability_Multiplier'] = (
    ((df['Academic Pressure'] + df['Financial Stress']) / 10) *
    (1 + 
     df['Have you ever had suicidal thoughts ?'] * 2.0 +
     df['Family History of Mental Illness'] * 1.5 +
     ((df['Sleep Duration'] <= 2).astype(int)) * 1.0 +
     ((df['Study Satisfaction'] <= 2).astype(int)) * 0.5
    )
)
````

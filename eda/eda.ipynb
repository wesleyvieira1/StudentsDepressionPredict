Limpeza dos dados

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df = df.drop(['id','Job Satisfaction','Work Pressure','Profession','City'], axis = 1)
df = df.dropna(subset=['Financial Stress'])
```

```python
df = df[df['Sleep Duration'] != 'Others']
df = df[df['Dietary Habits'] != 'Others']
df = df[df['Degree'] != 'Others']
```

Encoder

```python
def sleep_quality(sleep):
    if sleep == 'Less than 5 hours':
        return 1
    elif sleep == '5-6 hours':
        return 2
    elif sleep == '7-8 hours':
        return 3
    elif sleep == 'More than 8 hours':
        return 4
    else:
        return np.nan

df['Sleep Duration'] = df['Sleep Duration'].apply(sleep_quality)
```

```python
def dietary_habits(habit):
    if habit == 'Unhealthy':
        return 1
    if habit == 'Moderate':
        return 2
    if habit == 'Healthy':
        return 3
    else:
        return np.nan

df['Dietary Habits'] = df['Dietary Habits'].apply(dietary_habits)
df['Dietary Habits'].value_counts()
```

```python
def age_group(age):
    if 18 <= age < 25:
        return 1
    if 25 <= age < 35:
        return 2
    if 35 <= age < 50:
        return 3
    else:
        return 4

df['Age'] = df['Age'].apply(age_group)
df['Age'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Female':1,'Male': 2})
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map({'Yes':1,'No': 0})
df['Family History of Mental Illness']= df['Family History of Mental Illness'].map({'Yes':1,'No':0})
```

1. **Class 12** (Ensino Médio ou equivalente)
2. **B.Ed**, **B.Com**, **B.Arch**, **BCA**, **B.Tech**, **BHM**, **BSc**, **B.Pharm**, **BBA**, **BE**, **BA** (Graduação - Bacharelado)
3. **MSc**, **MCA**, **M.Tech**, **M.Ed**, **M.Com**, **MBA**, **MA**, **M.Pharm**, **MHM**, **ME** (Pós-Graduação - Mestrado)
4. **MBBS**, **MD** (Medicina - Graduação e Pós-Graduação)
5. **LLB**, **LLM** (Direito - Graduação e Pós-Graduação)
6. **PhD** (Doutorado - Nível mais alto de educação acadêmica)

```python
def superity_level(degree):
    if degree == 'Class 12':
        return 1
    elif degree in ['B.Ed', 'B.Com', 'B.Arch', 'BCA', 'B.Tech', 'BHM', 'BSc', 'B.Pharm', 'BBA', 'BE', 'BA']:
        return 2
    elif degree in ['MSc', 'MCA', 'M.Tech', 'M.Ed', 'M.Com', 'MBA', 'MA', 'M.Pharm', 'MHM', 'ME']:
        return 3
    elif degree in ['MBBS', 'MD']:
        return 4
    elif degree in ['LLB', 'LLM']:
        return 5
    elif degree == 'PhD':
        return 6
    else:
        return np.nan

df['Degree'] = df['Degree'].apply(superity_level)
```

Feature Eng

Índice de Estresse Geral

```python
# Definir os pesos para cada variável
weight_academic = 0.6  # Peso para Academic Pressure
weight_financial = 0.4  # Peso para Financial Stress

# Criar a nova métrica ponderada
df['Weighted Stress'] = (df['Academic Pressure'] * weight_academic + df['Financial Stress'] * weight_financial) / (weight_academic + weight_financial)
```

Qualidade de Vida

```python
# Definir os pesos para cada variável
weight_sleep = 0.4  # Peso para Sleep Duration
weight_diet = 0.3   # Peso para Dietary Habits
weight_study = 0.3  # Peso para Study Satisfaction

# Criar a nova métrica ponderada
df['Quality of Life'] = (
    (
        df['Sleep Duration'] * weight_sleep +
        df['Dietary Habits'] * weight_diet +
        df['Study Satisfaction'] * weight_study
    ) / (weight_sleep + weight_diet + weight_study)
).round(1)
```

Horas de Trabalho/Estudo por Resultado

```python
# Criar a nova feature "Efficiency"
df['Efficiency'] = df['CGPA'] / (df['Work/Study Hours'] + 1).round(1)  # Adiciona 1 para evitar divisão por zero
```

Histórico Familiar e Suicídio

```python
df['Risk Group'] = ((df['Family History of Mental Illness'] == 1) & (df['Have you ever had suicidal thoughts ?'] == 1)).astype(int)
df['Risk Group'].value_counts()
```

Identificar padrões de estresse.

```python
df['Academic_Work_Interaction'] = df['Academic Pressure'] * df['Work/Study Hours']
df['Academic_Financial_Interaction'] = df['Academic Pressure'] * df['Financial Stress']
df['Work_Financial_Interaction'] = df['Work/Study Hours'] * df['Financial Stress']
```

Interação entre Study Satisfaction e CGPA

```python
df['Study_CGPA_Interaction'] = df['Study Satisfaction'] * df['CGPA']
```

```python
corr_matrix = df.corr()
high_corr = corr_matrix[abs(corr_matrix) > 0.5]

plt.figure(figsize=(12, 10))
sns.heatmap(high_corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title('Matriz de Correlação')
plt.show()
```
